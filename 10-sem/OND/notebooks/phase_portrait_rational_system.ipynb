# Фазові портрети Cистеми

**Система** (область визначення $\mathcal D=\{(x,y): x+y>0\}$):

$$\dot x = P(x,y)=\frac{axy}{x+y}-a,\qquad
\dot y = Q(x,y)=-\frac{axy}{x+y}+by^2.$$

Параметри $a,b\in\mathbb R$. Позначимо $h(x,y)=\dfrac{xy}{x+y}$. Тоді
$\dot x=a(h-1)$, $\dot y=-ah+by^2$ і **$\dot x+\dot y=-a+by^2$**.


## 1. Точки рівноваги

Шукаємо $P=Q=0$ в $\mathcal D$.

### 1.1. Випадок $a=0$

$\dot x\equiv 0$, $\dot y=by^2$.

| $b$ | Рівноваги в $\mathcal D$ | Динаміка |
|-----|---------------------------|----------|
| $b\neq 0$ | пряма $y=0$, $x>0$ | $x(t)\equiv\mathrm{const}$, рух лише по $y$ |
| $b=0$ | уся $\mathcal D$ | тривіальна рівновага |

Якобіан на $y=0$ **вироджений** — класична лінеаризація не застосовна.

### 1.2. Випадок $a\neq 0$, $b=0$

З $Q=0$: $h=0 \Rightarrow xy=0$. Разом з $h=1$ із $P=0$ ізольованих точок **немає**.

### 1.3. Випадок $a\neq 0$, $b\neq 0$

З $P=0$: $h=1$, тобто
$$(x-1)(y-1)=1,\qquad y\neq 1,\qquad x=\frac{y}{y-1}.$$

З $Q=0$ при $h=1$: $by^2=a$, отже
$$y_\ast=\sqrt{\frac{a}{b}}\quad (потрібно $ab\ge 0$).$$

(Гілка $y=-\sqrt{a/b}$ дає $x+y=y^2/(y-1)<0$ при $y<0$ — **завжди поза $\mathcal D$**.)

**Умова перебування в $\mathcal D$** для $y_\ast>0$:
$$x+y=\frac{y^2}{y-1}>0 \iff y>1 \iff \begin{cases} b>0:\ a>b>0,\\ b<0:\ |a|>|b|. \end{cases}$$

Ізольована рівновага в $\mathcal D$:
$$E_\ast=\left(\frac{y_\ast}{y_\ast-1},\,y_\ast\right),\qquad y_\ast=\sqrt{a/b}.$$

**Приклад:** $a=4b$ $\Rightarrow$ $E_\ast=(2,2)$.

**Біфуркація входу в область:** при фіксованому $b>0$ точка з’являється в $\mathcal D$, коли $a\downarrow b^+$ ($y_\ast\to 1^+$, $x+y\to 0^+$).


## 2. Якобіан та власні значення

### 2.1. Загальний вигляд

$h_x=\dfrac{y^2}{(x+y)^2}$, $h_y=\dfrac{x^2}{(x+y)^2}$:

$$J(x,y)=\begin{pmatrix}
\dfrac{ay^2}{(x+y)^2} & \dfrac{ax^2}{(x+y)^2} \\[6pt]
-\dfrac{ay^2}{(x+y)^2} & -\dfrac{ax^2}{(x+y)^2}+2by
\end{pmatrix}.$$

### 2.2. В точці $E_\ast$ ($h=1$, $by_\ast^2=a$, $x+y=xy$)

Після підстановки $x_\ast=y_\ast/(y_\ast-1)$:

$$\mathrm{tr}\,J=a,\qquad
\det J=\frac{2a^2(y_\ast-1)^2}{y_\ast^3}>0\quad (y_\ast>1\text{ в }\mathcal D).$$

$$\Delta=\mathrm{tr}^2-4\det J=a^2\Bigl(1-\frac{8(y_\ast-1)^2}{y_\ast^3}\Bigr).$$

При $\Delta<0$: $\lambda=\dfrac{a}{2}\pm i\omega$ (фокус); при $\Delta>0$ — вузол; при $\Delta=0$ — вироджений вузол.

### 2.3. Класифікація в $\mathcal D$

У $\mathcal D$ при $a\neq 0$ **сідло неможливе** ($\det J>0$).

| Умова | Тип |
|--------|-----|
| $1<y_\ast<8$ | фокус ($\Delta<0$) |
| $y_\ast>8$ | вузол ($\Delta>0$) |
| $y_\ast=8$ | вироджений вузол ($\Delta=0$) |
| $a>0$ | нестійкий вузол/фокус |
| $a<0$ | стійкий вузол/фокус |

Узгоджена формула детермінанта в $E_\ast$:
$$\det J=\frac{2a^2(y_\ast-1)^2}{y_\ast^3},\qquad
\Delta=a^2-\frac{8a^2(y_\ast-1)^2}{y_\ast^3}=a^2\Bigl(1-\frac{8(y_\ast-1)^2}{y_\ast^3}\Bigr).$$

Фокус ($\Delta<0$) при $y_\ast^3>8(y_\ast-1)^2$ (наприклад $y_\ast=10$); при $y_\ast=2$, $a=4b$ маємо $\Delta=0$ — **вироджений вузол** ($\lambda=a/2$ кратності $2$), а не фокус.

$\mathrm{Re}\,\lambda=a/2\neq 0$ при $a\neq 0$ — **лінійний центр у ізольованій точці неможливий**.


## 3. Центр–фокус, біфуркація Хопфа, критерій Бендіксона

### 3.1. Проблема «центр–фокус»

Чисто уявні $\lambda=\pm i\omega$ потребують $\mathrm{tr}\,J=0$ і $\det J>0$. Оскільки $\mathrm{tr}\,J=a$, для $a\neq 0$ лінеаризація дає **фокус або вузол**, не центр.

Справжня ситуація центр–фокус: **$a=0$** — континуум рівноваг $y=0$, $x>0$.

### 3.2. Біфуркація Хопфа

Класичний Хопф ($\mathrm{tr}=0$, перетин уявної осі) **не реалізується** при зміні лише $b$ (слід $=a$). При $a\to 0$ — структурна зміна.

### 3.3. Критерій Бендіксона (раціональна система)

**Дивергенція** на $\mathcal D$ ($x+y>0$):

$$\mathrm{div}\,\mathbf F
= \frac{\partial P}{\partial x}+\frac{\partial Q}{\partial y}
= \frac{a(y-x)+2by^2(x+y)}{x+y}.$$

**Критерій Бендіксона:** якщо в однозв’язній області $\mathrm{div}\,\mathbf F$ **не змінює знак** (і не зникає), то в цій області **немає замкнених траєкторій**, що цілком лежать у ній.

Чисельник $N(x,y)=a(y-x)+2by^2(x+y)$ **змінює знак** на $\mathcal D$, тому на **всій** $\mathcal D$ критерій **не застосовується** — глобально цикли ним не виключаються. Для окремих підобластей $\mathcal D$, де $N$ одного знаку, висновок можливий **локально** (див. чисельну перевірку в коді).

### 3.4. Ізокліни ($a\neq 0$)

- $\dot x=0$: $(x-1)(y-1)=1$;
- $\dot y=0$: $y=0$ або $x=\dfrac{by^2}{a-by}$.


## 4. Візуалізація

Нижче — код у стилі `lab2/phase_portraits_3_6.py`: `solve_ivp`, `streamplot`, ізокліни, межа $x+y=0$.


```python
# =============================================================================
# Cистема: x' = a*xy/(x+y) - a,  y' = -a*xy/(x+y) + b*y^2
# Область D: x + y > 0.  Модуль візуалізації та класифікації рівноваг.
# =============================================================================

from __future__ import annotations

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Відступ від межі x+y=0: інтегратор зупиняється до виходу з області визначення
EPS = 1e-2


def _mask_domain(x: np.ndarray, y: np.ndarray, eps: float = EPS) -> np.ndarray:
    """Булева маска точок, де x+y > eps (тобто в області D)."""
    return (np.asarray(x) + np.asarray(y)) > eps


def h_xy(x: np.ndarray, y: np.ndarray) -> np.ndarray:
    """Спільний раціональний член h = xy/(x+y); поза D — NaN."""
    s = np.asarray(x) + np.asarray(y)
    with np.errstate(divide="ignore", invalid="ignore"):
        out = np.asarray(x, dtype=float) * np.asarray(y, dtype=float) / s
    # На самій прямій x+y=0 та поза D не будуємо поле
    out = np.where(np.abs(s) > 1e-14, out, np.nan)
    out = np.where(_mask_domain(x, y), out, np.nan)
    return out


def P(x, y, a: float) -> np.ndarray:
    """Компонента x' = a*(h - 1)."""
    return np.where(_mask_domain(x, y), a * (h_xy(x, y) - 1.0), np.nan)


def Q(x, y, a: float, b: float) -> np.ndarray:
    """Компонента y' = -a*h + b*y^2."""
    return np.where(_mask_domain(x, y), -a * h_xy(x, y) + b * np.asarray(y, dtype=float) ** 2, np.nan)


def equilibrium(a: float, b: float) -> np.ndarray | None:
    """
    Ізольована рівновага в D за аналітикою: y* = sqrt(a/b), x* = y*/(y*-1).
    Повертає None, якщо рівноваги в D немає (випадки a=0, b=0, ab<0, y*<=1).
    """
    if a == 0 or b == 0:
        return None
    if a * b < 0:
        return None
    y_star = float(np.sqrt(a / b))
    # У D потрібно y* > 1 (еквівалентно a>b при b>0)
    if y_star <= 1.0 + 1e-12:
        return None
    x_star = y_star / (y_star - 1.0)
    if x_star + y_star <= EPS:
        return None
    return np.array([x_star, y_star], dtype=float)


def jacobian_at(x: float, y: float, a: float, b: float) -> np.ndarray:
    """Матриця Якобі (аналітичні похідні від h = xy/(x+y))."""
    s = x + y
    if abs(s) < 1e-14:
        return np.full((2, 2), np.nan)
    s2 = s * s
    j11 = a * y * y / s2
    j12 = a * x * x / s2
    j21 = -a * y * y / s2
    j22 = -a * x * x / s2 + 2.0 * b * y
    return np.array([[j11, j12], [j21, j22]], dtype=float)


def classify(a: float, b: float) -> dict:
    """
    Локальний тип ізольованої рівноваги за tr(J), det(J), дискримінантом.
    У D det>0 — сідла немає; знак tr=a визначає стійкість.
    """
    if a == 0:
        if b == 0:
            return {"kind": "degenerate", "label": "a=b=0: уся D рівноважна"}
        return {"kind": "line", "label": "a=0: інваріантна пряма y=0, x>0"}
    if b == 0:
        return {"kind": "none", "label": "b=0: ізольованих рівноваг немає"}
    pt = equilibrium(a, b)
    if pt is None:
        return {"kind": "none", "label": "немає ізольованої рівноваги в D"}
    x, y = pt
    J = jacobian_at(x, y, a, b)
    tr = float(np.trace(J))
    det = float(np.linalg.det(J))
    disc = tr * tr - 4.0 * det  # <0 — фокус, >0 — вузол, ≈0 — вироджений
    y_star = y
    if abs(disc) < 1e-8 * max(1.0, abs(tr) ** 2):
        ltype = "вироджений вузол"
    elif disc < -1e-10:
        ltype = "фокус"
    else:
        ltype = "вузол"
    stab = "стійкий" if tr < 0 else "нестійкий" if tr > 0 else "нейтральний"
    lam = np.linalg.eigvals(J)
    return {
        "kind": "isolated",
        "point": pt,
        "jacobian": J,
        "trace": tr,
        "det": det,
        "discriminant": disc,
        "eigenvalues": lam,
        "label": f"{stab} {ltype}",
        "y_star": y_star,
    }


def make_rhs(a: float, b: float):
    """Повертає праву частину ODE для solve_ivp: z = [x, y]."""

    def rhs(t, z):
        x, y = z
        # Захист інтегратора біля межі D
        if x + y <= EPS:
            return [0.0, 0.0]
        s = x + y
        hxy = x * y / s
        return [a * (hxy - 1.0), -a * hxy + b * y * y]

    return rhs


def integrate_traj(a, b, x0, y0, t_span=(0.0, 12.0), max_step=0.05):
    """
    Одна траєкторія; event зупиняє інтегрування, коли x+y падає до EPS
    (вихід з області D).
    """
    rhs = make_rhs(a, b)

    def event_boundary(t, z):
        return z[0] + z[1] - EPS

    event_boundary.terminal = True
    event_boundary.direction = -1  # спрацьовує лише при перетині зсереду назовні

    sol = solve_ivp(
        rhs, t_span, [x0, y0],
        max_step=max_step, rtol=1e-8, atol=1e-10,
        events=event_boundary,
    )
    if not sol.success or sol.y.shape[1] < 2:
        return np.array([]), np.array([])
    return sol.y[0], sol.y[1]


def plot_nullclines(ax, a, b, xlim, ylim):
    """Криві P=0 (синя) та Q=0 (помаранчева) методом рівневих ліній."""
    nx, ny = 400, 400
    xg = np.linspace(xlim[0], xlim[1], nx)
    yg = np.linspace(ylim[0], ylim[1], ny)
    X, Y = np.meshgrid(xg, yg)
    Zp, Zq = P(X, Y, a), Q(X, Y, a, b)
    ax.contour(X, Y, Zp, levels=[0], colors="tab:blue", linewidths=1.6)
    ax.contour(X, Y, Zq, levels=[0], colors="tab:orange", linewidths=1.6)
    ax.plot([], [], color="tab:blue", label=r"$\dot x=0$")
    ax.plot([], [], color="tab:orange", label=r"$\dot y=0$")


def plot_domain_boundary(ax, xlim, ylim):
    """Межа D: пряма x+y=0; заштриховано заборонену півплощину x+y<=0."""
    xs = np.linspace(xlim[0], xlim[1], 200)
    ys = -xs
    mask = (ys >= ylim[0]) & (ys <= ylim[1])
    ax.plot(xs[mask], ys[mask], ":", color="dimgray", linewidth=1.4, label=r"$x+y=0$")
    verts_x = [xlim[0], xlim[1], xlim[1], xlim[0]]
    verts_y = [ylim[0], ylim[1], -xlim[1], -xlim[0]]
    ax.fill(verts_x, verts_y, color="lightgray", alpha=0.35, zorder=0)


def make_seeds(xlim, ylim, n=7, avoid=None):
    """
    Початкові точки для streamplot: сітка в D, без точок близько до avoid
    (зазвичай рівновага).
    """
    xs = np.linspace(xlim[0] + 0.1, xlim[1] - 0.1, n)
    ys = np.linspace(ylim[0] + 0.1, ylim[1] - 0.1, n)
    pts = []
    for x in xs:
        for y in ys:
            if x + y <= EPS + 0.05:
                continue
            ok = True
            if avoid:
                for c, r in avoid:
                    if np.hypot(x - c[0], y - c[1]) < r:
                        ok = False
                        break
            if ok:
                pts.append([x, y])
    return np.array(pts, dtype=float) if pts else np.zeros((0, 2))


def plot_phase_portrait(
    a: float,
    b: float,
    *,
    xlim=(-0.5, 5.0),
    ylim=(-1.0, 5.0),
    title: str | None = None,
    ax=None,
    t_forward=14.0,
    t_backward=-14.0,
):
    """
    Повний фазовий портрет для заданих (a,b):
    межа D, ізокліни, streamplot, чисельні траєкторії, маркер E*.
    """
    info = classify(a, b)
    if ax is None:
        fig, ax = plt.subplots(figsize=(8, 6.5), dpi=130)
    else:
        fig = ax.figure

    plot_domain_boundary(ax, xlim, ylim)
    if a != 0:
        plot_nullclines(ax, a, b, xlim, ylim)

    # Поле напрямків (лише там, де P,Q скінченні — через NaN поза D)
    xg = np.linspace(xlim[0], xlim[1], 55)
    yg = np.linspace(ylim[0], ylim[1], 55)
    X, Y = np.meshgrid(xg, yg)
    U, V = P(X, Y, a), Q(X, Y, a, b)

    avoid = []
    pt = equilibrium(a, b)
    if pt is not None:
        avoid.append((pt, 0.25))
    seeds = make_seeds(xlim, ylim, n=8, avoid=avoid if avoid else None)
    if seeds.size:
        ax.streamplot(X, Y, U, V, color="royalblue", density=1.2,
                      linewidth=1.4, arrowsize=1.1, start_points=seeds)

    # Чисельні траєкторії вперед і назад у часі
    ic_list = [
        (1.0, 0.5), (0.8, 2.0), (3.0, 3.5), (4.0, 0.8), (0.5, 3.0),
    ]
    if pt is not None:
        ic_list.extend([
            (pt[0] + 0.4, pt[1]), (pt[0], pt[1] + 0.5),
            (pt[0] - 0.3, pt[1] - 0.3),
        ])
    for x0, y0 in ic_list:
        if x0 + y0 <= EPS + 0.05:
            continue
        for tspan in ((0, t_forward), (0, t_backward)):
            xs, ys = integrate_traj(a, b, x0, y0, t_span=tspan)
            if xs.size:
                ax.plot(xs, ys, color="darkgreen", linewidth=1.2, alpha=0.85)

    # Окремий випадок a=0: континуум рівноваг на y=0
    if a == 0 and b != 0:
        ax.plot([xlim[0] + 0.05, xlim[1] - 0.05], [0, 0], "o", color="crimson",
                markersize=4, label="рівноваги: $y=0$, $x>0$")
    elif pt is not None:
        ax.plot(pt[0], pt[1], "o", color="crimson", markersize=10, zorder=5,
                label=rf"$E_\ast$ ({info['label']})")

    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_aspect("equal", adjustable="box")
    ax.grid(True, alpha=0.3)
    ax.axhline(0, color="black", linewidth=0.5)
    ax.axvline(0, color="black", linewidth=0.5)
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    ttl = title or rf"$a={a}$, $b={b}$: {info['label']}"
    ax.set_title(ttl, fontsize=11, fontweight="bold")
    ax.legend(loc="best", fontsize=8)
    return fig, ax, info

# =============================================================================
# Критерій Бендіксона — раціональна система (символьно + чисельна перевірка)
# =============================================================================

def divergence_rational(x: np.ndarray, y: np.ndarray, a: float, b: float) -> np.ndarray:
    """div F = (a(y-x) + 2*b*y^2*(x+y)) / (x+y) на D; поза D — nan."""
    s = np.asarray(x) + np.asarray(y)
    num = a * (np.asarray(y) - np.asarray(x)) + 2.0 * b * (np.asarray(y) ** 2) * s
    with np.errstate(divide="ignore", invalid="ignore"):
        out = num / s
    out = np.where(_mask_domain(x, y), out, np.nan)
    return out


def print_bendixson_rational(a: float, b: float, n_sample: int = 5) -> None:
    """
    Виводить div F для раціональної системи та оцінює знак на сітці в D.
    Якщо на сітці є обидва знаки — класичний Бендіксон на всій D не застосовний.
    """
    print("=" * 60)
    print(f"Бендіксон (раціональна система): a={a}, b={b}")
    print("  div F = (a(y-x) + 2*b*y^2*(x+y)) / (x+y)  на x+y>0")
    xs = np.linspace(0.3, 4.0, n_sample)
    ys = np.linspace(0.3, 4.0, n_sample)
    vals = []
    for xv in xs:
        for yv in ys:
            if xv + yv <= EPS:
                continue
            d = float(divergence_rational(xv, yv, a, b))
            vals.append(d)
            print(f"    ({xv:.2f}, {yv:.2f})  ->  div = {d:+.4f}")
    if vals:
        has_pos = any(v > 0 for v in vals)
        has_neg = any(v < 0 for v in vals)
        if has_pos and has_neg:
            print("  Висновок: знак div НЕ постійний -> критерій Бендіксона на всій D не спрацьовує.")
        elif has_pos:
            print("  На обраній сітці div > 0 (локально можна виключити цикли лише в підобласті цього знаку).")
        else:
            print("  На обраній сітці div < 0 (аналогічно для від'ємного знаку).")
    print()


# Демонстрація для типового режиму (можна викликати з іншими a, b)
print_bendixson_rational(9.0, 1.0)
print_bendixson_rational(4.0, 1.0)

```

## 5. Висновки

1. В $\mathcal D$ існує **щонайбільше одна** ізольована рівновага при $ab>0$ і $y_\ast=\sqrt{a/b}>1$.
2. У $\mathcal D$ **немає сідел** ($\det J>0$); можливі лише вузол або фокус.
3. **Бендіксон (раціональна система):** $\mathrm{div}\,\mathbf F$ не має постійного знаку на всій $\mathcal D$ — глобально цикли не виключені.
4. **Приклад 3:** на $y>-\tfrac12$ маємо $\mathrm{div}>0$, на $y<-\tfrac12$ — $\mathrm{div}<0$; у кожній півплощині Бендіксон виключає цикли, що цілком у ній лежать.
5. **Приклад 4:** $\mathrm{div}\,\mathbf F$ змінює знак на площині; у прямокутнику навколо $O$ — теж; локально $O$ — стійкий фокус (лінійна теорія).

## Приклад 4. Система $\dot x=(x+\frac73)(\frac12 y-x)$, $\dot y=(y-\frac43)(x+y)$

### Рівноваги

$$O(0,0),\quad A(-\tfrac73,\tfrac43),\quad B(-\tfrac73,\tfrac73),\quad C(\tfrac23,\tfrac43).$$

### Ізокліни

- $\dot x=0$: $x=-\dfrac73$ або $y=2x$;
- $\dot y=0$: $y=\dfrac43$ або $y=-x$.

Прямі $x=-\dfrac73$, $y=\dfrac43$ — інваріантні.

| Точка | Тип |
|-------|-----|
| $O$ | стійкий фокус |
| $A$, $C$ | сідло |
| $B$ | нестійкий вузол |

### Критерій Бендіксона

$$\mathrm{div}\,\mathbf F = -x+\frac{5}{2}y-\frac{11}{3}.$$

Множина $\mathrm{div}=0$ — пряма; на $\mathbb R^2$ знак **не постійний**, тому **глобально** Бендіксон не виключає всі цикли.

У прямокутнику «без контактів» навколо $O$ $\mathrm{div}$ **також змінює знак** — критерій на весь прямокутник не застосовується.

**Локально в $O$:** $\mathrm{div}(O)=-\tfrac{11}{3}<0$, стійкий фокус — **ізольованого циклу** в малому околі $O$ немає (лінійна теорія).


```python
# %pip install sympy numpy scipy matplotlib
#
# Символьна перевірка аналітики для прикладів 3 і 4 з навчального ескізу.

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

sp.init_printing()

# =============================================================================
# Приклад 3: x' = x, y' = y^2  (сідло-вузол у (0,0))
# =============================================================================
print("=" * 72)
print("ПРИКЛАД 3:  x' = x,  y' = y^2")
print("=" * 72)

x, y, t, C = sp.symbols("x y t C", real=True)
xd3 = x
yd3 = y**2

# --- Рівноваги: P=Q=0 ---
eq_pts3 = sp.solve([xd3, yd3], [x, y], dict=True)
print("Рівноваги:", eq_pts3)

# --- Якобіан J та власні значення в (0,0) ---
J3 = sp.Matrix([[sp.diff(xd3, x), sp.diff(xd3, y)],
                [sp.diff(yd3, x), sp.diff(yd3, y)]])
print("J(x,y) =")
sp.pprint(J3)
J3_0 = J3.subs({x: 0, y: 0})
print("J(0,0) =", J3_0)
print("Власні значення в (0,0):", J3_0.eigenvals())  # 1 та 0 → сідло-вузол


# --- Вірний інтеграл (з dy/dx = y^2/x) ---
H_right = x * sp.exp(1 / y)
dH_right = sp.simplify(sp.diff(H_right, x) * xd3 + sp.diff(H_right, y) * yd3)
print("Вірно:   x*exp(1/y);   d/dt =", dH_right)

# Еквівалент: ln|x| + 1/y = C (на x≠0, y≠0)
H_log = sp.log(sp.Abs(x)) + 1 / y
dH_log = sp.simplify(sp.diff(H_log, x) * xd3 + sp.diff(H_log, y) * yd3)
print("Також:   ln|x| + 1/y;  d/dt =", dH_log)

print("\nФазове рівняння: dy/dx =", sp.simplify(yd3 / xd3))
print("Розділення: dx/x = dy/y^2  =>  ln|x| + 1/y = C")

# =============================================================================
# Приклад 4: поліноміальна система, 4 рівноваги
# =============================================================================
print("\n" + "=" * 72)
print("ПРИКЛАД 4")
print("=" * 72)

xd4 = (x + sp.Rational(7, 3)) * (y / 2 - x)
yd4 = (y - sp.Rational(4, 3)) * (x + y)

eqs4 = sp.solve([xd4, yd4], [x, y], dict=True)
print("Рівноваги (sympy):", eqs4)

# Факторизація показує ізокліни: x=-7/3, y=2x, y=4/3, y=-x
print("\nІзокліни x'=0:", sp.factor(sp.expand(xd4)))
print("Ізокліни y'=0:", sp.factor(sp.expand(yd4)))

# Чи інваріантні прямі x=-7/3 та y=4/3 (вектор швидкості вздовж них)?
print("\nНа x=-7/3:  x' =", sp.simplify(xd4.subs(x, -sp.Rational(7, 3))))
print("На y=4/3:   y' =", sp.simplify(yd4.subs(y, sp.Rational(4, 3))))

J4 = sp.Matrix([[sp.diff(xd4, x), sp.diff(xd4, y)],
                [sp.diff(yd4, x), sp.diff(yd4, y)]])

labels4 = {
    "O": (0, 0),
    "A": (-sp.Rational(7, 3), sp.Rational(4, 3)),
    "B": (-sp.Rational(7, 3), sp.Rational(7, 3)),
    "C": (sp.Rational(2, 3), sp.Rational(4, 3)),
}

for name, pt in labels4.items():
    Jp = J4.subs({x: pt[0], y: pt[1]})
    ev = Jp.eigenvals()
    print(f"\n{name}{pt}: tr={sp.simplify(Jp.trace())}, det={sp.simplify(Jp.det())}")
    print("  eigenvalues:", list(ev.keys()))

# =============================================================================
# Критерій Бендіксона — усі три системи (sympy)
# =============================================================================
print("\n" + "=" * 72)
print("КРИТЕРІЙ БЕНДІКСОНА")
print("=" * 72)

# --- Раціональна система ---
a_sym, b_sym = sp.symbols("a b", real=True)
h_sym = x * y / (x + y)
P_rat = a_sym * (h_sym - 1)
Q_rat = -a_sym * h_sym + b_sym * y**2
div_rat = sp.simplify(sp.diff(P_rat, x) + sp.diff(P_rat, y))
print("\n[Раціональна] div F =", div_rat)
print("  Критерій: потрібен постійний знак div на однозв'язній області.")
print("  На всій D знак не постійний (див. чисельний вивід print_bendixson_rational).")

# --- Приклад 3 ---
div3 = sp.simplify(sp.diff(xd3, x) + sp.diff(yd3, y))
print("\n[Приклад 3] div F =", div3)
print("  Нуль div: y =", sp.solve(div3, y))
print("  y > -1/2  =>  div > 0  =>  немає циклів у підобластях цієї півплощини")
print("  y < -1/2  =>  div < 0  =>  немає циклів у підобластях нижньої півплощини")

# --- Приклад 4 ---
div4 = sp.simplify(sp.diff(xd4, x) + sp.diff(yd4, y))
print("\n[Приклад 4] div F =", div4)
print("  Нуль div:", sp.Eq(div4, 0))
print("  div(O) =", sp.simplify(div4.subs({x: 0, y: 0})))
for name, pt in labels4.items():
    print(f"  div({name}) =", sp.simplify(div4.subs({x: pt[0], y: pt[1]})))
print("  Знак div на площині не постійний -> глобально Бендіксон не виключає всі цикли")
print("  O — стійкий фокус: ізольованого циклу в малому околі O немає")

```

```python
# =============================================================================
# Візуалізація прикладів 3 і 4 (два підграфіки в одній фігурі)
# =============================================================================

def integrate_gen(rhs, x0, y0, t_end=8.0, n_eval=400):
    """
    Інтегрування від (x0,y0) вперед і назад у часі.
    dense_output=True дає гладку криву на рівномірній сітці tt.
    """
    curves = []
    for sign in (1.0, -1.0):
        sol = solve_ivp(
            rhs,
            (0.0, sign * t_end),
            [x0, y0],
            rtol=1e-9,
            atol=1e-11,
            max_step=0.03,
            dense_output=True,
        )
        if not sol.success:
            continue
        tt = np.linspace(0.0, sign * t_end, n_eval)
        xx, yy = sol.sol(tt)
        curves.append((xx, yy))
    return curves


# ------------------------- Приклад 3 -------------------------
def rhs3(t, z):
    """Права частина: x'=x, y'=y^2."""
    return [z[0], z[1] ** 2]


def plot_example3(ax):
    xlim, ylim = (-2.2, 2.2), (-2.2, 2.2)

    # Поле напрямків
    X, Y = np.meshgrid(np.linspace(xlim[0], xlim[1], 30),
                       np.linspace(ylim[0], ylim[1], 30))
    U, V = X, Y**2
    ax.streamplot(X, Y, U, V, color="royalblue", density=1.1, linewidth=1.0, arrowsize=1.0)

    # Аналітичні траєкторії: рівні H(x,y)=x*exp(1/y)=C  =>  y = 1/ln(C/x)
    xg = np.linspace(-2.0, -0.05, 200)   # x<0
    xg_p = np.linspace(0.05, 2.0, 200)   # x>0
    for Cval in [-2.0, -0.5, 0.5, 1.0, 2.0, 4.0]:
        for xs in (xg, xg_p):
            with np.errstate(divide="ignore", invalid="ignore"):
                ratio = Cval / xs
            valid = ratio > 0  # ln(C/x) визначений
            ln_r = np.log(ratio[valid])
            yc = 1.0 / ln_r
            m = (yc > ylim[0]) & (yc < ylim[1]) & np.isfinite(yc)
            if np.any(m):
                ax.plot(xs[valid][m], yc[m], color="darkgreen", linewidth=1.3, alpha=0.85)

    # Інваріантна вісь Ox: y=0, x'=x (стрілки від/до нуля)
    ax.axhline(0, color="tab:red", linewidth=1.2, label=r"$y=0$ (інваріантна)")
    ax.axvline(0, color="black", linewidth=0.5)
    for x0 in np.linspace(-1.8, 1.8, 9):
        if abs(x0) < 0.08:
            continue
        dx = 0.25 * np.sign(x0)
        ax.annotate("", xy=(x0 + dx, 0), xytext=(x0, 0),
                    arrowprops=dict(arrowstyle="->", color="tab:red", lw=1.2))

    # Додаткові чисельні траєкторії (сідло-вузол)
    ics = [(-1.0, -0.6), (-0.4, -0.3), (0.4, 0.35), (1.0, 0.6), (0.3, -0.5)]
    for x0, y0 in ics:
        for xx, yy in integrate_gen(rhs3, x0, y0, t_end=6.0):
            ax.plot(xx, yy, color="purple", linewidth=1.1, alpha=0.75)

    ax.plot(0, 0, "ko", markersize=9, zorder=6, label=r"$(0,0)$ сідло-вузол")
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_aspect("equal", adjustable="box")
    ax.grid(True, alpha=0.3)
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    ax.set_title(r"Приклад 3: $\dot x=x$, $\dot y=y^2$; інтеграл $x e^{1/y}=C$", fontweight="bold")
    ax.legend(loc="upper left", fontsize=8)


# ------------------------- Приклад 4 -------------------------
def rhs4(t, z):
    x, y = z
    return [
        (x + 7.0 / 3.0) * (0.5 * y - x),
        (y - 4.0 / 3.0) * (x + y),
    ]


def plot_example4(ax):
    xlim, ylim = (-3.5, 2.5), (-0.5, 3.5)

    # streamplot на грубій сітці
    X, Y = np.meshgrid(np.linspace(xlim[0], xlim[1], 28),
                       np.linspace(ylim[0], ylim[1], 28))
    U = (X + 7.0 / 3.0) * (0.5 * Y - X)
    V = (Y - 4.0 / 3.0) * (X + Y)
    ax.streamplot(X, Y, U, V, color="royalblue", density=1.0, linewidth=0.9, arrowsize=0.9)

    # Нулькліни (точніша сітка для contour)
    xg = np.linspace(xlim[0], xlim[1], 400)
    yg = np.linspace(ylim[0], ylim[1], 400)
    Xf, Yf = np.meshgrid(xg, yg)
    Zx = (Xf + 7.0 / 3.0) * (0.5 * Yf - Xf)
    Zy = (Yf - 4.0 / 3.0) * (Xf + Yf)
    ax.contour(Xf, Yf, Zx, levels=[0], colors="tab:blue", linewidths=1.4)
    ax.contour(Xf, Yf, Zy, levels=[0], colors="tab:orange", linewidths=1.4)
    ax.plot([], [], color="tab:blue", label=r"$\dot x=0$")
    ax.plot([], [], color="tab:orange", label=r"$\dot y=0$")

    # Чотири рівноваги з ескізу
    eqs = {
        "O": (0.0, 0.0),
        "A": (-7.0 / 3.0, 4.0 / 3.0),
        "B": (-7.0 / 3.0, 7.0 / 3.0),
        "C": (2.0 / 3.0, 4.0 / 3.0),
    }
    colors_eq = {"O": "crimson", "A": "darkgreen", "B": "purple", "C": "chocolate"}
    for name, (xe, ye) in eqs.items():
        ax.plot(xe, ye, "o", color=colors_eq[name], markersize=9, zorder=5, label=f"${name}$")

    # Спіралі, що наближаються до стійкого фокуса O
    spiral_ics = [(1.2, 1.0), (0.8, 2.2), (-0.5, 2.5), (1.5, 0.5), (-1.0, 1.2)]
    for x0, y0 in spiral_ics:
        sol = solve_ivp(rhs4, (0, 25), [x0, y0], rtol=1e-9, atol=1e-11, max_step=0.05)
        if sol.success:
            ax.plot(sol.y[0], sol.y[1], color="darkgreen", linewidth=1.2, alpha=0.85)

    # Прямокутник «без контактів» навколо O (траєкторії входять всередину)
    rect_x = [-0.2, 1.8, 1.8, -0.2, -0.2]
    rect_y = [0.3, 0.3, 2.7, 2.7, 0.3]
    ax.plot(rect_x, rect_y, "k--", linewidth=1.5, label="область без контактів")
    ax.annotate("спіралі до $O$", xy=(0.15, 0.2), fontsize=9, color="darkgreen")

    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_aspect("equal", adjustable="box")
    ax.grid(True, alpha=0.3)
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    ax.set_title(
        "Приклад 4: "
        + r"$\dot x=(x+\frac{7}{3})(\frac{y}{2}-x)$, "
        + r"$\dot y=(y-\frac{4}{3})(x+y)$",
        fontweight="bold",
        fontsize=10,
    )
    ax.legend(loc="upper right", fontsize=7)


# --- Збірка фігури ---
fig, axes = plt.subplots(1, 2, figsize=(15, 6.2), dpi=130)
plot_example3(axes[0])
plot_example4(axes[1])
fig.suptitle("Приклади 3 і 4 — фазові портрети", fontsize=13, fontweight="bold")
fig.tight_layout()
plt.show()
```
